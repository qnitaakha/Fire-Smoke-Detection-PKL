> **Result**

```python
import pandas as pd

def get_results(csv_path, model_name):
    df = pd.read_csv(csv_path).tail(1)
    return {
        "Model": model_name,
        "Precision": float(df['metrics/precision(B)']),
        "Recall": float(df['metrics/recall(B)']),
        "mAP50": float(df['metrics/mAP50(B)']),
        "mAP50-95": float(df['metrics/mAP50-95(B)'])
    }

compare_df = pd.DataFrame([
    get_results(f"/kaggle/working/runs/detect/fire_detection_base{i}/results.csv", f"Baseline {i}")
    for i in [1, 2, 3]
])

compare_df
```

```python
from ultralytics import YOLO
import pandas as pd

eval_configs = [
    {
        "name": "Baseline 1 - Test",
        "weights": "/kaggle/input/fire-and-smoke-detection-model/pytorch/default/1/fire_detection_base1/weights/best.pt",
        "data": "/kaggle/input/dataset1/data.yaml"
    },
    {
        "name": "Baseline 2 - Test",
        "weights": "/kaggle/input/fire-and-smoke-detection-model/pytorch/default/1/fire_detection_base2/weights/best.pt",
        "data": "/kaggle/input/dataset3/data.yaml"
    },
    {
        "name": "Baseline 3 - Test",
        "weights": "/kaggle/input/fire-and-smoke-detection-model/pytorch/default/1/fire_detection_base3/weights/best.pt",
        "data": "/kaggle/input/dataset4/data.yaml"
    },
]

rows = []

for cfg in eval_configs:
    print(f"Evaluating: {cfg['name']}")
    
    model = YOLO(cfg["weights"])
    metrics = model.val(
        data=cfg["data"],
        split="test",
        verbose=False    # boleh True kalau mau lihat log detail
    )
    
    results = metrics.results_dict
    
    rows.append({
        "Model": cfg["name"],
        "Precision": results["metrics/precision(B)"],
        "Recall": results["metrics/recall(B)"],
        "mAP50": results["metrics/mAP50(B)"],
        "mAP50-95": results["metrics/mAP50-95(B)"]
    })

test_results_df = pd.DataFrame(rows)
test_results_df
```
